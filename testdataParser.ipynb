

# this can parse all RegionID,CarType,NumberDays,NumberSeats,StartTimestamp,EndTimestamp,Price,HasVollkasko,FreeKilometers

```python
# this will parse all different values of RegionID,CarType,NumberDays
import json

regionIDs = []
carTypes = []
numberDays = []

#reading the testdata.json file
testdata = "testdata.json"

testdataJson = json.load(open(testdata))

for i in range(0,len(testdataJson)):
    regionIDs.append(testdataJson[i]['RegionID'])
    carTypes.append(testdataJson[i]['CarType'])
    numberDays.append(testdataJson[i]['NumberDays'])


regionIDs = list(set(regionIDs))
carTypes = list(set(carTypes))
numberDays = list(set(numberDays))

#exporting all to a text file:
#first line is the number of different values of RegionID
#second line is the number of different values of CarType
#third line is the number of different values of NumberDays
#then all the different values of RegionID
#then all the different values of CarType
#then all the different values of NumberDays

with open('testdata.txt', 'wb') as f:
    stringToAppend = ''
    stringToAppend += str(len(regionIDs)) + '\n'
    stringToAppend += str(len(carTypes)) + '\n'
    stringToAppend += str(len(numberDays)) + '\n'
    for i in range(0,len(regionIDs)):
        stringToAppend += str(regionIDs[i]) + '\n'
    for i in range(0,len(carTypes)):
        stringToAppend += str(carTypes[i]) + '\n'
    for i in range(0,len(numberDays)):
        stringToAppend += str(numberDays[i]) + '\n'
    print(stringToAppend)

    
```
